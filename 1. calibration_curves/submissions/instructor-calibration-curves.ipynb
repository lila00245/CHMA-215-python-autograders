# OH MY that took a while, how can we make that go faster ??

Did you notice that there was only one difference in finding each response value? The only change was the index value of the response being subtracted from. We can use a 'for loop' to automate this process with A LOT less lines of code.

A for loop looks like the following:
```
for i in range(length_of_list):   # where i is the index value you're currently on and goes from i=0 to i=n(length_of_list)-1
  do something                    # for loop logic, can be one line of code, can be multiple
```

If you take the example list above: **tv_shows = ["squid games", "gossip girl", "avatar the last airbender", "american dad"]**, we can use a for loop to iterate or loop over each value in the list.

```
tv_shows = ["squid games", "gossip girl", "avatar the last airbender", "american dad"]

for i in range(len(tv_shows)):  
  print(tv_shows[i])
```

This code prints each tv show in the list on a separate line starting with squid games and ending with american dad.

* **len(tv_shows)** returns the length of the list, in this case there are 4 items, so the for loop would run 4 times, starting with i=0 and ending with i=3 (because n=4 and it goes to n-1).
* **print(tv_shows[i])** uses **i** as the index variable to print the tv show at each index value.
* **do something** can include one line of code like in this example, or it can include a lot more depending on what you want to be looped over


````python
# these four lines of code do the exact same as the thirteen lines above
corrected_responses2 = []

# looping through our data_length (0 to n-1) where n=6 data items
for i in range(data_length):
  corrected = response[i] - response[0]    # calculating corrected value
  corrected_responses2.append(corrected)   # adding corrected value to list
````

````python
# 1. creating the scatter plot
# change the color of the dots / data points by swapping 'black' for '<color of your choice>'

# TODO FOR STUDENT
plt.scatter(x_array, y_array, color="black")

# 2. calculating the variables for the trendline
# the lineregress function takes a set of paired data sets (x_array, y_array) and
# returns the slope (m), y-intercept (b), r-value (r), p-value (p) and standard error (stderr)
# we have to assign all values it returns to a variable, but we won't be using the last two
m, b, r, p, stderr = stats.linregress(x_array, y_array)

# 3. plotting the trendline
# change the color of the trendline by swapping 'red' for '<color of your choice>'
# mess around with the other customization options laid out above, make the graph unique

# TODO FOR STUDENT
plt.plot(
    x_array, m*x_array + b,          # x and y values for trendline, don't change this

    # --- customizable --- #
    color='red',                     # color of trendline
    linestyle='--',                  # line style: '--', '-.', ':', '-'
    linewidth=2                      # line width
)

# 4. adding trendline and r^2 equations to graph
# f"..." : f-string, lets us put variables directly inside the text using {var}
# \n : new line in the text
trendline_string = (
    f"y = {m:.3f}x + {b:.3f}\n"   # {var:.3f} : uses value of var and shows first 3 decimal places
    f"R$^2$ = {r**2:.4f}"         # R$^2$ : LaTeX math so Matplotlib displays R² nicely, {r**2:.4f} : R^2 with 4 decimal places
)
# adds the trendline equation text to the plot at a fixed position
plt.annotate(trendline_string, xy=(0.14, 0.71), xycoords="figure fraction")  # position is 14% from the left and 71% from the bottom of the *entire figure*

# 5. customizing graph

# TODO FOR STUDENT
plt.title("Calibration Curves Scatter Plot", fontsize=16, fontweight='bold')               # add title of graph
plt.xlabel("Concentration [μM]", fontsize=12)                                              # label x axis
plt.ylabel("Detector Response [mV]", fontsize=12)                                          # label y axis

# axis limits and gridlines customization
plt.grid(True, linestyle=':', linewidth=0.7, alpha=0.7)
plt.xlim(0, 25)
plt.ylim(0, 800)

# 6. showing scatter plot
plt.show()

# printing full slope and y-intercept values
print("slope =", m)
print("y-intercept =", b)
````

# Having fun yet?

Going back to our unknown detector response sample (mV), how do we find the concentration (μM) using the trendline we just created?

$$y = m*x + b$$

We have the slope (m), y-intercept (b), and our unknown detector response sample (y). Using algebra, we are going to find the concentration (x) for that sample.

````python
# rearrange the trendline equation to find the concentration (x)
# remember to use corrected_unk for the known y value
# don't forget to use () and check that the value you find makes sense in relation to your graph

# TODO FOR STUDENT
x = (corrected_unk - b) / m   # finding x value using x = (y-b)/m

# prints the corrected unknown (y) and concentration value (x)
print("corrected_unk (y) =", corrected_unk)
print("x =", x)
````
